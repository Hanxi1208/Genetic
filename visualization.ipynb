```python
import plotly.express as px

fig = px.scatter_matrix(df, dimensions=df.columns[:-1], color="Fitness")
# fig = px.scatter_matrix(df, dimensions=df.columns[:], color="Fitness")
fig.update_layout(width=1500, height=1500, title="Scatter Matrix View")

fig.show()
```

```python
import plotly.express as px
fig = px.parallel_coordinates(df, color="Fitness", labels={"InnerRadius":"InnerRadius", "Length":"Length", 
                                                           "OpeningAngle":"OpeningAngle", "InnerRadius2":"InnerRadius2",
                                                           "Length2":"Length2","OpeningAngle2":"OpeningAngle2", 
                                                           "Generation":"Generation","Fitness":"Fitness",},
                    color_continuous_scale=px.colors.diverging.Tealrose, color_continuous_midpoint=4.6)
fig.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(16, 8))
gs = gridspec.GridSpec(2, 4)

ax0 = plt.subplot(gs[0])
sns.distplot(df['InnerRadius'], bins = 20)

ax1 = plt.subplot(gs[1])
sns.distplot(df['Length'], bins = 20)

ax2 = plt.subplot(gs[2])
sns.distplot(df['OpeningAngle'], bins = 20)

ax3 = plt.subplot(gs[3])
sns.distplot(df['InnerRadius2'], bins = 20)

ax4 = plt.subplot(gs[4])
sns.distplot(df['Length2'], bins = 20)

ax5 = plt.subplot(gs[5])
sns.distplot(df['OpeningAngle2'], bins = 20)

ax6 = plt.subplot(gs[6])
sns.distplot(df['Generation'], bins = 20)

ax7 = plt.subplot(gs[7])
sns.distplot(df['Fitness'], bins = 20)

plt.tight_layout()
plt.show()
```
